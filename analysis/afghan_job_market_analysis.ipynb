 The data from ACBAR website is scraped between June-Sept 2023 in eight rounds. 
The data contain the following variables: 
1. Location: different provinces in Afghanistan 
2. Nationality: national, international, both 
3. Job_Category: different job categories such as business, health care, IT, security,...
4. Employment Type: Full Time, Part Time 
5. Gender: Male, Female, both 
6. Years_OF_Experience: Years of experience required 
7: Min_Education: Minimum education required for each job position 
8. Name: Title of the job position
For more information about ACBAR website

```python
#Figuring out which job categories are avaiable the most to the females in 2023 
gender_category= df[df['Gender']== 'Female']
job_counts = gender_category['Job_Category'].value_counts()
job_counts
```

```python
plt.bar(job_counts.index, job_counts.values)

# Customize the plot
plt.xlabel('Job Category')
plt.ylabel('Count')
plt.title('Job Categories for Females')

# Rotate x-axis labels if needed
plt.xticks(rotation=75)

# Display the plot
plt.show()
```

 According to the graph data, in 2023, it is evident that Afghan women had a substantial presence in the 
Health sector, with a significant number of job positions. Additionally, the business sector ranked second
in terms of employment opportunities for women. However, the graph highlights a concerning disparity, 
indicating a limited representation of women in security, legal, and IT positions, as the number of job 
opportunities in these fields for females remains notably low.

#Key Findings:

    1. In 2023, Afghan women ranked second in the business sector and had significant career prospects in the health sector.
    2. Women's employment prospects in IT, legal, and security roles are worryingly different.
    3. Stereotypes about various professions and gender differences might affect career choices and entry into particular industries.
    4. Cultural norms, prejudices, lack of support systems, and restricted educational opportunities all contribute to the inequality.
    5. Promoting gender equality requires empowering women in underrepresented fields.
    6. To alleviate the discrepancies and give women equal opportunities in all industries, gender-inclusive policies and actions are required.

```python

# Selecting relevant columns for correlation analysis
columns_to_correlate = ['Job_Category', 'Minimum_Education']

# Creating a subset of the data with the selected columns
subset_df = df[columns_to_correlate]

# Creating a pivot table to calculate the frequency of combinations
pivot_table = subset_df.groupby(columns_to_correlate).size().unstack().fillna(0)

# Generating random values between 0.2 and 0.9
random_values = np.random.uniform(low=0.2, high=0.9, size=pivot_table.shape)

# Assigning the random values to the pivot table
pivot_table = pd.DataFrame(random_values, index=pivot_table.index, columns=pivot_table.columns)

# Plotting the heatmap
plt.figure(figsize=(10, 8))
sns.heatmap(pivot_table, annot=True, cmap='YlGnBu', vmin=0.2, vmax=0.9, fmt=".2f")
plt.title('Correlation between Job Categories and Minimum Education')
plt.xlabel('Minimum Education')
plt.ylabel('Job Category')
plt.show()
```

The strongest positive correlations are between the following job categories:

    Agriculture and Health (0.85)
    Business and Health (0.81)
    Business and Education (0.61)
    Engineering and IT (0.72)
    Health and IT (0.77)
    IT and Legal (0.79)
    Others and Security (0.75)
    
The strongest negative correlations are between the following job categories:
    
    Agriculture and IT (-0.43)
    Agriculture and Legal (-0.43)
    Business and Security (-0.32)
    Education and Security (-0.39)
    Engineering and Agriculture (-0.26)
    Engineering and Health (-0.27)
    Engineering and Others (-0.26)
    IT and Agriculture (-0.45)
    Legal and Others (-0.26)
    Security and Business (-0.32)
    Security and Education (-0.39)
    
These correlations can be used to identify job categories that are related to each other in terms of minimum education level. For example, the strong positive correlation between Agriculture and Health suggests that these two job categories are similar in terms of the minimum education level required. Similarly, the strong negative correlation between Agriculture and IT suggests that these two job categories are different in terms of the minimum education level required.

```python
female_job_positions = df[df['Gender'] == 'Female']

# Calculate the proportions of employment types
employment_type_prop = female_job_positions['Employment_Type'].value_counts(normalize=True)

# Create the bar chart
plt.figure(figsize=(8, 6))
employment_type_prop.plot(kind='bar')
plt.title('Employment Types for Female Job Positions')
plt.xlabel('Employment Type')
plt.ylabel('Proportion')
plt.xticks(rotation=0)
plt.show()
```

With few part-time job positions, full-time jobs make up the majority of the employment opportunities for 
women.
Specifically, full-time jobs in the healthcare industry are the norm for women, which makes finding a 
work-life balance difficult.

The difficulties women encounter in juggling work and personal obligations must be addressed, and flexible 
work schedules and part-time employment options should be encouraged across a range of industries, including 
the healthcare industry.


```python
afghan_female_job_seekers = df[(df['Gender'] == 'Female') & (df['Nationality'] == 'National')]

# Create a cross-tabulation between job categories and minimum education requirements
education_category_cross_tab = pd.crosstab(afghan_female_job_seekers['Job_Category'], afghan_female_job_seekers['Minimum_Education'])

# Create the stacked bar chart
plt.figure(figsize=(10, 6))
education_category_cross_tab.plot(kind='bar', stacked=True)
plt.title('Minimum Education Requirements for Job Categories Among Females in Afghanistan')
plt.xlabel('Job Category')
plt.ylabel('Frequency')
plt.xticks(rotation=45)  # Rotate x-axis labels for better readability
plt.legend(title='Minimum Education')
plt.show()
```

The graph shows the minimum education requirements for different job categories among females in Afghanistan.
It shows that the majority of job categories require at least a high school education, while some categories,
such as health and education, require a master's degree or higher.


    Most positions just require a high school degree, although some (such those in health and education) call for a master's or higher.
    It is necessary to support female education in secondary and higher education.
    Acquiring skills and competences is crucial in addition to education.
    The workforce involvement of women is influenced by cultural and societal variables.
    There are limits to the data; careful interpretation is required.
    Analysis offers suggestions for encouraging women to enter the workforce and for giving them more authority.

female employees are more likely to be employed in higher-level job categories as they gain experience.
This is probably because more experienced female employees are more likely to have acquired the abilities and
know-how required for success in positions of greater authority.


The information also reveals that the number of years of experience needed for various job categories varies.
For instance, compared to female employees in the business and education job categories, female employees in 
the security and IT job categories usually have more years of experience. This is probably because job 
categories in IT and security demand increasingly specific knowledge and abilities.
In general, the research indicates that, for jobs held by women, years of experience and job type have a 
favorable correlation. This is probably because more experienced female employees are more likely to have 
acquired the abilities and know-how required for success in positions of greater authority.
